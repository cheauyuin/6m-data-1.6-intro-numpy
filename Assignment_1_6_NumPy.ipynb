### 3. When to use `float` vs. `int` in real analyst work

In real-world data analysis, the choice between `float` and `int` depends on the nature of the data and the precision required:

*   **Integers (`int`)**:
    *   **Counts**: Any data representing discrete counts of items, people, or events should be integers (e.g., number of customers, product inventory, transaction count).
    *   **Identifiers**: Unique IDs, year values, or categorical encodings are typically integers (e.g., `customer_id`, `product_sku`, `zip_code`).
    *   **Exact values**: When fractional parts are meaningless or misleading.

*   **Floats (`float`)**:
    *   **Measurements**: Any continuous measurement that can have fractional values (e.g., temperature, weight, height, distance).
    *   **Currency/Prices**: While prices are often displayed with two decimal places, they are inherently fractional and should be stored as floats to allow for calculations and accurate representation (e.g., `$19.99`).
    *   **Percentages/Ratios**: Values between 0 and 1 or representing proportions (e.g., `0.75` for 75%).
    *   **Calculations resulting in fractions**: If intermediate or final calculations might produce non-whole numbers, floats are necessary to maintain precision.

Choosing the correct data type is crucial for accuracy, memory efficiency, and preventing unexpected behavior in analyses.

I would prefer using float when decimal precision is required. For example, price, markings, height/weight, and a result from a calculation that might return fraction instead of whole number.

On the other hand, I would prefer integer when it only involves whole number. For example, age, year, unique numeric ID, or counts (no. of employees, quantity of items).

# 3. Boolean masks: Work with targets
**Scenario:** You have campaign response data.

```python
names = np.array(["Ana", "Ben", "Chen", "Dana", "Eli", "Fatima", "George", "Hui"])
spend = np.array([200, 150, 300, 120, 180, 220, 160, 310])      # marketing spend
revenue = np.array([400, 180, 500, 100, 220, 260, 150, 600])   # revenue
```

1. Compute the **ROI** for each person: roi = revenue / spend.

```python
roi = revenue / spend
print("ROI for each person:")
print(roi)
```

2. Create a boolean mask for customers with roi >= 2.0.

```python
roi_mask = roi >= 2.0
print("Boolean mask for ROI >= 2.0:")
print(roi_mask)
```

3. Use the mask to:

    *  List their names
    *  List their spend and revenue

```python
print("Names of customers with ROI >= 2.0:")
print(names[roi_mask])

print("Spend of customers with ROI >= 2.0:")
print(spend[roi_mask])

print("Revenue of customers with ROI >= 2.0:")
print(revenue[roi_mask])
```

4. Create a second mask for customers with spend >= 200.

```python
spend_mask = spend >= 200
print("Boolean mask for Spend >= 200:")
print(spend_mask)
```

5. Combine the two masks to find customers who have **roi >= 2.0 AND spend >= 200**.

```python

combined_mask = (roi_mask & spend_mask)
print("Customers with ROI >= 2.0 AND Spend >= 200:")
print(f"Names: {names[combined_mask]}")
print(f"Spend: {spend[combined_mask]}")
print(f"Revenue: {revenue[combined_mask]}")
```

# 4. Gradebook: Aggregations and broadcasting
Reuse the “Gradebook” idea from class and extend it.

```python
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])
```

1. Compute each student’s **average score** (per row).


```python
average_scores = np.mean(scores, axis=1)
print("Each student's average score:")
print(average_scores)
```

2. Compute each subject’s **average score** (per column).

```python
average_subject_scores = np.mean(scores, axis=0)
print("Each subject's average score:")
print(average_subject_scores)
```

```python
# This is my study note

import numpy as np

# Basic usage
arr = np.array([1, 2, 3, 4])
average = np.mean(arr)  # Output: 2.5
print(average)

# Multi-dimensional usage
arr_2d = np.array([[1, 2], [3, 4]])
col_mean = np.mean(arr_2d, axis=0)  # Output: [2.0, 3.0]
print("\nColumn average: ")
print(col_mean)

row_mean = np.mean(arr_2d, axis=1)
print("\nRow average: ")
print(row_mean)
```

3. Create scores_centered by subtracting the **subject mean** from each column (broadcasting).

```python
scores_centered = scores - average_subject_scores
print("Scores Centered (subject mean subtracted from each column):")
print(scores_centered)
print(f"Shape: {scores_centered.shape}")
print(f"Data Type: {scores_centered.dtype}")
```

4. For scores_centered, compute per-student averages again.

```python
per_student_averages_centered = np.mean(scores_centered, axis=1)
print("Per-student averages from scores_centered:")
print(per_student_averages_centered)
```

5. Compare which student looks best by **raw average** vs **centered average**.

```python
per_student_averages_centered
```

```python
print("Raw Averages:")
for i, avg in enumerate(average_scores):
    print(f"Student {students[i]}: {avg:.2f}")

print("\nCentered Averages:")
for i, avg in enumerate(per_student_averages_centered):
    print(f"Student {students[i]}: {avg:.2f}")

# Identify best student by raw average
best_raw_avg_index = np.argmax(average_scores)
best_raw_avg_student = students[best_raw_avg_index]
print(f"\nStudent with best raw average: {best_raw_avg_student} (Score: {average_scores[best_raw_avg_index]:.2f})")

# Identify best student by centered average
best_centered_avg_index = np.argmax(per_student_averages_centered)
best_centered_avg_student = students[best_centered_avg_index]
print(f"Student with best centered average: {best_centered_avg_student} (Score: {per_student_averages_centered[best_centered_avg_index]:.2f})")
```

```python
import numpy as np


X = np.array([
    [10,  3.5,  0],
    [25,  5.0,  1],
    [40,  2.0,  0],
    [15, 10.0,  3],
    [30,  4.0,  2]
])
customers = np.array(["C1", "C2", "C3", "C4", "C5"])

# Weight set 1: Balanced
w1 = np.array([0.1, 0.5, 1.0])
scores1 = X @ w1

# Weight set 2: Heavy on Purchases
w2 = np.array([0.5, 1.0, 5.0])
scores2 = X @ w2

print("Balanced Rankings:", customers[np.argsort(-scores1)])
print("Purchase-heavy Rankings:", customers[np.argsort(-scores2)])
```

### Random Test Scores Generation and Analysis

```python
import numpy as np

# Generate 1,000 random test scores with np.random.randn
np.random.seed(42) # for reproducibility
raw_scores = np.random.randn(1000)

# Scale them to have mean 70 and standard deviation 10
mean_target = 70
std_dev_target = 10
scaled_scores = raw_scores * std_dev_target + mean_target

# Clip scores to the range [0, 100]
clipped_scores = np.clip(scaled_scores, 0, 100)

print("--- Clipped Scores Statistics ---")
print(f"Min Score: {np.min(clipped_scores):.2f}")
print(f"Max Score: {np.max(clipped_scores):.2f}")
print(f"Mean Score: {np.mean(clipped_scores):.2f}")
print(f"Standard Deviation: {np.std(clipped_scores):.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(clipped_scores, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Clipped Test Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### A/B Test Simulation

```python
# Simulate a small A/B test

# Group A: 20 users
n_users_a = 20
# Randomly generate conversions (0 or 1) for Group A
# Let's assume a baseline conversion rate for Group A, e.g., 40%
np.random.seed(101) # for reproducibility
conversions_a = np.random.choice([0, 1], size=n_users_a, p=[0.6, 0.4])

# Group B: 20 users
n_users_b = 20
# Randomly generate conversions (0 or 1) for Group B
# Let's assume an improved conversion rate for Group B, e.g., 55%
np.random.seed(102) # for reproducibility
conversions_b = np.random.choice([0, 1], size=n_users_b, p=[0.45, 0.55])

# Compute conversion rate per group using pure NumPy operations
conversion_rate_a = np.mean(conversions_a)
conversion_rate_b = np.mean(conversions_b)

print("--- A/B Test Results ---")
print(f"Group A Conversions: {conversions_a}")
print(f"Group A Conversion Rate: {conversion_rate_a:.2%}")
print(f"\nGroup B Conversions: {conversions_b}")
print(f"Group B Conversion Rate: {conversion_rate_b:.2%}")

# A simple comparison
if conversion_rate_b > conversion_rate_a:
    print(f"\nGroup B performed better with a conversion rate of {conversion_rate_b:.2%}")
elif conversion_rate_a > conversion_rate_b:
    print(f"\nGroup A performed better with a conversion rate of {conversion_rate_a:.2%}")
else:
    print("\nBoth groups performed equally.")
```
