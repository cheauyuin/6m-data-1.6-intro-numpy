# 2. Sales table drill: Indexing and slicing
**Scenario**: You have quarterly sales numbers for 4 regions (rows) over 5 quarters (columns).

```python
import numpy as np

sales = np.array([
    [120, 135, 110, 145, 130],  # Region A
    [ 90,  85,  95, 100, 110],  # Region B
    [200, 215, 190, 225, 245],  # Region C
    [155, 140, 135, 160, 170]   # Region D
])
regions = np.array(["A", "B", "C", "D"])
quarters = np.array(["Q1", "Q2", "Q3", "Q4", "Q5"])
```

Do the following:

1. Select all quarters for **Region B** as a 1D array.
2. Select **Q2 to Q4** (inclusive) for all regions as a 2D subarray.
3. Select **Q5** sales for **Regions A and D only** (use slicing or fancy indexing).
4. Compute the **shape** of each result and add a brief comment: “Is this 1D or 2D, and why?”

```python
sales
```

```python
# 1. Select all quarters for Region B as a 1D array.

print("Sales for Region B:")
region_b_sales = sales[1, :] # (index 1) for Region B
print(region_b_sales)
print(f"Shape: {region_b_sales.shape}")
print("Comment: This is a 1D array because we selected a single row (Region B), collapsing the dimension of regions. It represents a sequence of sales values for that specific region.\n")
```

```python
# 2. Select Q2 to Q4 (inclusive) for all regions as a 2D subarray.

print("Sales for Q2 to Q4 across all regions:")
quarters_q2_q4_sales = sales[:, 1:4]  # Q2 (index 1) to Q4 (index 3, exclusive for stop index, so it includes Q4)
print(quarters_q2_q4_sales)
print(f"Shape: {quarters_q2_q4_sales.shape}")
print("Comment: This is a 2D array because we selected multiple rows (all regions) and multiple columns (Q2 to Q4), preserving both the region and quarter dimensions.\n")
```

```python
# 3.Select Q5 sales for Regions A and D only (use slicing or fancy indexing).

print("Q5 Sales for Regions A and D:")
q5_regions_ad_sales = sales[[0, 3], 4]  # Regions A (index 0) and D (index 3), Q5 (index 4)
print(q5_regions_ad_sales)
print(f"Shape: {q5_regions_ad_sales.shape}")
print("Comment: This is a 1D array because we selected specific rows (Regions A and D) and a single column (Q5), effectively flattening the result into a list of sales values for the chosen regions in Q5.")
```

# 3. Boolean masks: Work with targets
**Scenario:** You have campaign response data.

```python
names = np.array(["Ana", "Ben", "Chen", "Dana", "Eli", "Fatima", "George", "Hui"])
spend = np.array([200, 150, 300, 120, 180, 220, 160, 310])      # marketing spend
revenue = np.array([400, 180, 500, 100, 220, 260, 150, 600])   # revenue
```

1. Compute the **ROI** for each person: roi = revenue / spend.

```python
roi = revenue / spend
print("ROI for each person:")
print(roi)
```

2. Create a boolean mask for customers with roi >= 2.0.

```python
roi_mask = roi >= 2.0
print("Boolean mask for ROI >= 2.0:")
print(roi_mask)
```

3. Use the mask to:

    *  List their names
    *  List their spend and revenue

```python
print("Names of customers with ROI >= 2.0:")
print(names[roi_mask])

print("Spend of customers with ROI >= 2.0:")
print(spend[roi_mask])

print("Revenue of customers with ROI >= 2.0:")
print(revenue[roi_mask])
```

4. Create a second mask for customers with spend >= 200.

```python
spend_mask = spend >= 200
print("Boolean mask for Spend >= 200:")
print(spend_mask)
```

5. Combine the two masks to find customers who have **roi >= 2.0 AND spend >= 200**.

```python
combined_mask = (roi_mask & spend_mask)
print("Customers with ROI >= 2.0 AND Spend >= 200:")
print(f"Names: {names[combined_mask]}")
print(f"Spend: {spend[combined_mask]}")
print(f"Revenue: {revenue[combined_mask]}")
```

6. In a markdown cell, answer: “What business insight do you get from this filtered group?”

Answer:

These are high value customers who spend high amount and generate high return. The company should run campaign to retain these customers.

# 4. Gradebook: Aggregations and broadcasting
Reuse the “Gradebook” idea from class and extend it.

```python
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])
```

1. Compute each student’s **average score** (per row).

```python
student_average = scores.mean(axis=1)
print("Average scores per student:")
print(student_average)
```

2. Compute each subject’s **average score** (per column).

```python
subject_average = scores.mean(axis=0)
print("Average scores per subject:")
print(subject_average)
```

3. Create scores_centered by subtracting the **subject mean** from each column (broadcasting).

```python
scores_centered = scores - subject_average
print("Scores centered around subject means:")
print(scores_centered)
```

4. For scores_centered, compute per-student averages again.

```python
student_avg_centered = scores_centered.mean(axis=1)
print("Average scores per student after centering:")
print(student_avg_centered)
```

5. Compare which student looks best by **raw average** vs **centered average**.

```python
print(student_average)
print(student_avg_centered)
```

S3 looks best.

# 5. Reshape and flatten: From daily to weekly
**Scenario**: You have 30 days of daily website visits, and you want to summarize them by week.

```python
daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])
```

```python
daily_visits
```

1. Reshape `daily_visits` into a `(6, 5)` array representing **6 weeks × 5 days**.


```python
# Assumming the first 5 element forms a week
# Total 6 weeks

weekly_visits = daily_visits.reshape((6, 5))
print("Weekly Visits:")
print(weekly_visits)
```

2. Compute **total visits per week**.

```python
# Axis = 1 sum across all colimns
# Returning 1 sum per row (week)

total_visit_per_week = weekly_visits.sum(axis=1)
print("Total visits per week:")
print(total_visit_per_week)
```

3. Compute **average visits per day of the week** (e.g., all “day 1 of week” together, all “day 2 of week” together, etc.).

```python
avg_visit_day_of_week = weekly_visits.mean(axis=0)
print("Average visits per day of the week:")
print(avg_visit_day_of_week)
```

4. Flatten the reshaped array back to 1D and confirm it matches the original `daily_visits`.

```python
reshaped_to_1d = weekly_visits.flatten()
print("Reshaped back to 1D:")
print(reshaped_to_1d)

print("\nFlattened array equal to original? ")
np.array_equal(daily_visits, reshaped_to_1d)
```

```python
scores1 = X @ w1
print(scores1)
```

3. Rank customers by score (highest first).

```python
rank_by_score = customers[np.argsort(scores1)[::-1]]
print(rank_by_score)
```

```python
w2 = np.array([0.5, 1.0, 5.0])
print("Weight heavy on past_purchases", w2)

scores2 = X @ w2
print("Score heavy on past_purchases", scores2)

rank_by_score_purchase = customers[np.argsort(scores2)[::-1]]
print("Rank heavy on past_purchases")
print(rank_by_score_purchase)
```

### Random Test Scores Generation and Analysis

```python
import numpy as np

# Generate 1,000 random test scores with np.random.randn
np.random.seed(42) # for reproducibility
raw_scores = np.random.randn(1000)

# Scale them to have mean 70 and standard deviation 10
mean_target = 70
std_dev_target = 10
scaled_scores = raw_scores * std_dev_target + mean_target

# Clip scores to the range [0, 100]
clipped_scores = np.clip(scaled_scores, 0, 100)

print("--- Clipped Scores Statistics ---")
print(f"Min Score: {np.min(clipped_scores):.2f}")
print(f"Max Score: {np.max(clipped_scores):.2f}")
print(f"Mean Score: {np.mean(clipped_scores):.2f}")
print(f"Standard Deviation: {np.std(clipped_scores):.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(clipped_scores, bins=20, edgecolor='black', alpha=0.7)
plt.title('Distribution of Clipped Test Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

### A/B Test Simulation

```python
# Simulate a small A/B test

# Group A: 20 users
n_users_a = 20
# Randomly generate conversions (0 or 1) for Group A
# Let's assume a baseline conversion rate for Group A, e.g., 40%
np.random.seed(101) # for reproducibility
conversions_a = np.random.choice([0, 1], size=n_users_a, p=[0.6, 0.4])

# Group B: 20 users
n_users_b = 20
# Randomly generate conversions (0 or 1) for Group B
# Let's assume an improved conversion rate for Group B, e.g., 55%
np.random.seed(102) # for reproducibility
conversions_b = np.random.choice([0, 1], size=n_users_b, p=[0.45, 0.55])

# Compute conversion rate per group using pure NumPy operations
conversion_rate_a = np.mean(conversions_a)
conversion_rate_b = np.mean(conversions_b)

print("--- A/B Test Results ---")
print(f"Group A Conversions: {conversions_a}")
print(f"Group A Conversion Rate: {conversion_rate_a:.2%}")
print(f"\nGroup B Conversions: {conversions_b}")
print(f"Group B Conversion Rate: {conversion_rate_b:.2%}")

# A simple comparison
if conversion_rate_b > conversion_rate_a:
    print(f"\nGroup B performed better with a conversion rate of {conversion_rate_b:.2%}")
elif conversion_rate_a > conversion_rate_b:
    print(f"\nGroup A performed better with a conversion rate of {conversion_rate_a:.2%}")
else:
    print("\nBoth groups performed equally.")
```
